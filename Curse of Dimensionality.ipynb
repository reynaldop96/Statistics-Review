```python
import numpy as np
import random
import sklearn.metrics.pairwise
import scipy.spatial.distance
import matplotlib.pyplot as plt

n = 10000
std_dev= []
for dim in range(100):
    c = np.random.randint(1000, size = (n,dim))
    std = np.std(scipy.spatial.distance.pdist(c, 'euclidean'))
    std_dev.append(std)
    
plt.plot(std_dev[2:])
plt.ylabel('St Dev of all Pairwise Distances')
plt.xlabel('dimension')
plt.show()
```
